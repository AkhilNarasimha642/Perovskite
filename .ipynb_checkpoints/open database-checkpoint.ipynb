### regression

```python
seed = 21899
traindata,testdata = train_test_split(data_total,test_size=0.20,random_state=seed)
```

```python
traindata.head()
```

```python
feature_list = ['anion_X','A_X','B_X','anion_IE','A_IE','B_IE','density','A_s_R','A_p_R','A_d_R','A_aff','B_s_R','B_p_R','B_d_R','B_aff']
def compute_mse(n,feature_list,clf,y_train,y_test,norm):
    # n is the number of features that you want to use in KRR
    # norm=1 means do the normalization of x_train and y_train data
    x_train = [0.]*n
    x_test = [0.]*n
    for r in itertools.combinations(feature_list,n):
        for i in range(len(r)):
            x_train[i] = traindata[r[i]]
            x_test[i] = testdata[r[i]]
        x_train = np.transpose(x_train)
        x_test = np.transpose(x_test)
        if norm == 1:
            x_train_scaler = StandardScaler().fit(x_train)
            x_train = x_train_scaler.transform(x_train)
            x_test = x_train_scaler.transform(x_test)
        clf.fit(x_train,y_train)
        y_predict = clf.predict(x_test)
        mse = mean_squared_error(y_test,y_predict)
        Rsquared = clf.score(x_test,y_test)
        print(r)
        print('mse: '+str(mse))
        print('R-squared: '+str(Rsquared))
    return clf,y_predict
        
```

```python
def plot_predict(predict,y_test):
    plt.scatter(predict,y_test,alpha=0.2)
    x = np.arange(min(predict),max(predict),0.01)
    plt.plot(x,x,color='black')
    plt.plot(x,0.9*x,linestyle='--',color='red')
    plt.plot(x,1.1*x,linestyle='--',color='red')
    return
```

```python
def LinearRidge(n,feature_list,L1wt,y_train,y_test):
    # n is the number of features that you want to use in KRR
    x_train = [0.]*n
    x_test = [0.]*n
    for r in itertools.combinations(feature_list,n):
        for i in range(len(r)):
            x_train[i] = traindata[r[i]]
            x_test[i] = testdata[r[i]]
        x_train = np.transpose(x_train)
        x_test = np.transpose(x_test)
        mod = sm.OLS(y_train,x_train)
        res = mod.fit_regularized(L1_wt=L1wt)
        print(res.summary())
        print(res.params)
    return res
```

```python
def feature_select(features,n,corr):
    corrlist = []
    combine = []
    for r in itertools.combinations(features,n):
        sum_corr = 0.
        for i in itertools.combinations(r,2):
            sum_corr = sum_corr + corr[i[0]][i[1]]
        corrlist.append(sum_corr)
        combine.append([r])
    df = pd.DataFrame({'sum of correlation':corrlist,'combination':combine})
    return df
    
```

```python
from scipy import stats

def Ttest(outlier,pred_compare):
    sig_list = []
    p_list = []
    for x in list(outlier.columns):
        if x not in ['anion','A_ion','B_ion']:
            statistic,pvalue = stats.ttest_ind(pred_compare[x],outlier[x])
            if pvalue<0.05:
                sig_list.append(x)
                p_list.append(pvalue)
    return sig_list,p_list
    
```

```python
temp = pred_compare.sample(n=len(outlier_CB_ind))
temp_siglist,tempplist=Ttest(temp,pred_compare)
siglist_CB,plist_CB=Ttest(outlier_CB_ind,pred_compare)
```

```python
print(temp_siglist)
print(siglist_CB)
```

```python
temp = pred_compare.sample(n=len(outlier_VB_ind))
temp_siglist,tempplist=Ttest(temp,pred_compare)
siglist_VB,plist_VB=Ttest(outlier_VB_ind,pred_compare)
```
