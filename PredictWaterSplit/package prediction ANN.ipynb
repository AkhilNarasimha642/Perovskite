### This notebook is to show how we make prediction by Nerual Network

```python
from KRRcode import open_database
perovskite,values,data_total = open_database.read_database()
```

the dataframe 'data_total' is a collection of all the data

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import matplotlib

%matplotlib inline
```

```python
def model(): #define a neural network model
    model = Sequential()
    model.add(Dense(19, input_dim=19, kernel_initializer='normal', activation='relu'))
    model.add(Dense(12, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

```python
def plot_comparison(predict,y_test,title): #define a function for comparing predicted data and actual data
    plt.scatter(y_test,predict,alpha=0.2)
    x = np.arange(min(y_test),max(y_test),0.01)
    plt.plot(x,x,color='black',label='y = x')
    plt.plot(x,0.9*x,linestyle='--',color='red',alpha=0.5,label = 'y = 90% x')
    plt.plot(x,1.1*x,linestyle='--',color='red',alpha=0.5,label = 'y = 110% x')
    plt.xlabel('True Value')
    plt.ylabel('Predictive Value')
    plt.title(title)
    plt.legend()
    plt.savefig(title+'.png', dpi = 350)
    return
```

```python
seed = 21899
```

## direct valence band  ##

```python
X = data_total[['anion_X', 'anion_IE', 'A_X', 'A_IE', 'A_s_R',
       'A_p_R', 'A_d_R', 'A_aff' , 'B_X', 'B_IE', 'B_s_R', 'B_p_R',
       'B_d_R', 'B_aff', 'volume', 'mass', 'density', 'A_R', 'B_R']].values
Y = data_total[['VB_dir']].values
```

```python
X_train_pn,X_test_pn,y_train,y_test1 = train_test_split(X, Y, test_size=0.20,random_state=seed)
```

```python
X_train_scaler = StandardScaler().fit(X_train_pn)
X_train = X_train_scaler.transform(X_train_pn)
```

```python
np.random.seed(seed)
estimator = KerasRegressor(build_fn= model,
        epochs=500, batch_size=10000, verbose=0)
history = estimator.fit(X_train, y_train, validation_split=0.33, epochs=500, 
        batch_size=5000, verbose=0)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
X_test = X_train_scaler.transform(X_test_pn)
```

```python
test_loss = estimator.model.evaluate(X_test, y_test1)
test_loss # displaying the MSE value for testing data
```

```python
predictions_VB = estimator.model.predict(X_test,batch_size=30000)
predictions_VB  # derive the predicted value for testing data
```

```python
plot_comparison(predictions_VB,y_test1,'Prediction on valence band')
```
